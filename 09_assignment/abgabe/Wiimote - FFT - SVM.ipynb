```python
from pylab import *
from scipy import fft
from sklearn import svm
import numpy as np
import os
import csv
```

```python
# get current directory
curdir = os.path.dirname(os.path.realpath("__file__"))
# fill activity-arrays with csv-data
box_csv = [curdir+"/trainingdata/box1.csv", curdir+"/trainingdata/box2.csv", curdir+"/trainingdata/box3.csv", curdir+"/trainingdata/box4.csv", curdir+"/trainingdata/box5.csv", curdir+"/trainingdata/box6.csv", curdir+"/trainingdata/box7.csv", curdir+"/trainingdata/box8.csv", curdir+"/trainingdata/box9.csv", curdir+"/trainingdata/box10.csv", curdir+"/trainingdata/box11.csv", curdir+"/trainingdata/box12.csv"]
walk_csv = [curdir+"/trainingdata/walk1.csv", curdir+"/trainingdata/walk2.csv", curdir+"/trainingdata/walk3.csv", curdir+"/trainingdata/walk4.csv", curdir+"/trainingdata/walk5.csv", curdir+"/trainingdata/walk6.csv", curdir+"/trainingdata/walk7.csv", curdir+"/trainingdata/walk8.csv", curdir+"/trainingdata/walk9.csv", curdir+"/trainingdata/walk10.csv", curdir+"/trainingdata/walk11.csv", curdir+"/trainingdata/walk12.csv"]
hop_csv = [curdir+"/trainingdata/hop1.csv", curdir+"/trainingdata/hop2.csv", curdir+"/trainingdata/hop3.csv", curdir+"/trainingdata/hop4.csv", curdir+"/trainingdata/hop5.csv", curdir+"/trainingdata/hop6.csv", curdir+"/trainingdata/hop7.csv", curdir+"/trainingdata/hop8.csv", curdir+"/trainingdata/hop9.csv", curdir+"/trainingdata/hop10.csv", curdir+"/trainingdata/hop11.csv", curdir+"/trainingdata/hop12.csv"]
```

```python
def read_data(filename):
    x = []
    y = []
    z = []
    avg = []
    els = []
    for line in open(filename, "r").readlines():
        # check for proper file-structure
        if(len(line.strip().split(",")) == 3):
            _x, _y, _z = map(int,line.strip().split(","))
        x.append(_x)
        y.append(_y)
        z.append(_z)
        avg.append((_x+_y+_z)/3)

        """
        # precompute lists and append to list
        for el in (_x+_y+_z):
            # cast to float for more detailed results
            els.append(float(el) / 3)
        for num in els:
            # only append numbers of the list, not the whole lists
            avg.append(num)
        """
    return avg
```

```python
# fill raw activity array with csv-entries
box_raw = [read_data(f) for f in box_csv]
walk_raw  = [read_data(f) for f in walk_csv]
hop_raw  = [read_data(f) for f in hop_csv]
```

```python
all = box_raw + walk_raw + hop_raw
minlen = min([len(x) for x in all])
print "Cutting off after", minlen , "samples"
# cut off (normalize csv's by shortest column in all csv's)
box_cut = [l[:minlen] for l in box_raw]
walk_cut = [l[:minlen] for l in walk_raw]
hop_cut = [l[:minlen] for l in hop_raw]
```

```python
# calculate fft
print len(box_cut)
box_freq = [np.abs(fft(l)/len(l))[1:len(l)/2] for l in box_cut]
walk_freq = [np.abs(fft(l)/len(l))[1:len(l)/2] for l in walk_cut]
hop_freq = [np.abs(fft(l)/len(l))[1:len(l)/2] for l in hop_cut]
```

```python
# plot boxing-frequency
for x in range(8):
    plot(range(len(box_freq[x])), box_freq[x])
```

```python
# plot hopping-frequency
for x in range(8):
    plot(range(len(hop_freq[x])), hop_freq[x])
```

```python
# plot walking-frequency
for x in range(8):
    plot(range(len(walk_freq[x])), walk_freq[x])
```

```python
c = svm.SVC()
```

```python
BOX = 0
WALK = 1
HOP = 2
categories = [0] * 8 + [1] * 8 + [2] * 8
# setup training-data (limited to 3/4 of data)
training_data = box_freq[:8] + walk_freq[:8] + hop_freq[:8]
# train SVM
print len(training_data)
print '-------'
print len(categories)
c.fit(training_data,categories)
```

```python
# setup recognition array
activities = ["Box", "Walk", "Hop"]
```

```python
# predict activity for remaining boxing-data
pred_box = c.predict([box_freq[8], box_freq[9], box_freq[10], box_freq[11]])
for i in range(0, len(pred_box)):
    print str(i + 1) + ". Prediction (Box): " + str(activities[pred_box[i]])
```

```python
# predict activity for remaining hopping-data
pred_hop = c.predict([hop_freq[8], hop_freq[9], hop_freq[10], hop_freq[11]])
for i in range(0, len(pred_hop)):
    print str(i + 1) + ". Prediction (Hop): " + str(activities[pred_hop[i]])
```

```python
# predict activity for remaining walking-data
pred_walk = c.predict([walk_freq[8], walk_freq[9], walk_freq[10], walk_freq[11]])
for i in range(0, len(pred_walk)):
    print str(i + 1) + ". Prediction (Walk): " + str(activities[pred_walk[i]])
```

Documentation
=============
The experiment implements a support vector machine (SVM). The SVM was trained with data of three different activities, where each dataset was measured for 10 seconds per activity. The three activities are boxing, hopping and walking.
The wiimote was held inside a trouser pocket while walking and hopping. Boxing was executed while holding the wiimote in the right hand and alternately boxing straight forward with the right and left hand.
For each activity twelve samples were collected, eight for training the SVM and four to validate the SVM by calculating a prediction for them. The SVM predicts all remaining datasets correctly.
